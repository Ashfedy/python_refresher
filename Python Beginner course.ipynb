```python
#If...Else statement

#Syntax --> Indentation is important in the syntax after mentioning the condition

#if <condition>:
#----...
#elif <condition>:
#    ...
#else
#    ...

age = 15
if age >= 18:
    print("You're an adult")
elif age >=13:
    print("You're a teenager")
else:
    print("You're a kid")
```

```python
#For...in loop

#syntax

#for temp_variable in list_name:
#---...

for country in countries:
    print(country)
```

```python
#Using enumerate function in For...in loop

#syntax

#for enumerating_variable, temp_variable in enumerate(list_name):
#    ...

for i, country in enumerate(countries):
    print(i)
    print(country)
```

```python
#using for...in loop with dictionary

#syntax

#for key,value in dictionary_name.items():
#    ...

for key,value in dictionary.items():
    print(key)
    print(value)
```

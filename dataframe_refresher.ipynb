# Assigning multiple series to a dataframe
## Assign() --> Gives a copy of the dataframe and does not update the dataframe itself

```python
series1=pd.Series(np.random.randint(1,100,1000),index=np.arange(0,1000))
series2=pd.Series(np.random.uniform(1,100,1000),index=np.arange(0,1000))
```

```python
df_student_performance = df_student_performance.assign(Series_column_1=series1,Series_column_2=series2,)
```

```python
df_student_performance
```

## Insert() --> Updated the dataframe itself directly

```python
df_student_performance.insert(1,"rand_col_inserted",series1,allow_duplicates=True)
```

```python
df_student_performance
```

```python
#deletes all columns with the name "rand_col_inserted"
df_student_performance.drop(columns=["rand_col_inserted"])
```

# Dataframe

```python
import pandas as pd
import numpy as np
```

```python
# creating an array

# 1st way of creating an array
# data = np.array([["Ashwin",32],["Niveda",33],["Julia",23]])

# 2nd way of creating an array
data = [["Vedhanth",5],["Ashwin",32],["Niveda",33],["Jayanthi",59],["Julia",23],["Aravinth",32]]
```

```python
# creating a dataframe
df = pd.DataFrame(data,index=['1','2','3','4','5','6'],columns=['Name','Age'])
```

```python
# display dataframe
df
```
